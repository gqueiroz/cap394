# 1.1 Dados de entrada
<hr style="border:1px solid #0077b9;">

Para explorar as funcionalidades do GeoPandas serão utilizadas três bases de dados geográficos: focos de incêndio do ano de 2017, unidades federativas e muninípios do Brasil. Os dados podem ser econtrados no arquivo de descrição da aula (README.md).

![Dados](img/dados.png)

Após descompactar os arquivos os dados serão apresentados no formato shapefile. O shapefile é um formato de arquivo que contém dados geoespaciais (dados vetoriais) utilizados em Sistemas de Informação Geográfica (GIS). Os shapfiles das bases de dados são formados por arquivos com as seguintes extensões:

- .cpg é um arquivo que descreve a codificação aplicada para criar o shapefile.
- .dbf é um arquivo de banco de dados usado para armazenar dados de atributos e IDs de objetos.
- .prj é um arquivo que contém os metadados associados ao sistema de coordenadas e projeção shapefiles.
- .shp é um arquivo  que fornece aos recursos sua geometria. 
- .shx o arquivo de índice que armazena o índice da geometria do recurso.


Os arquivos shapfile podem ser carregados em um SIG como o QGis ou o TerraView, por exemplo. Neste caso, cada shapfile será carregado na forma de uma camada (ou mapa). Um mapa é uma represetação, em escala, de uma seleção de entidades abstratas relacionadas com a superfície da Terra. Na imagem abaixo é apresentada a visualização dos dados no TerraView, note que cada base de dados foi carregada em uma camada (canto superior esquerdo).

# 5. Operações espaciais 
<hr style="border:2px solid #0077b9;">

O GeoPandas disponibiliza um conjunto de operações para manipulação de objetos geográficos. Os tipos geográficos utilizados no GeoPandas utilizam a implementação do pacote shapely, desta forma, todas operações disponíveis pelo shapely também estão disponíveis no GeoPandas. Nesta seção serão apresentados alguns exemplos de operações espaciais, para mais detalhes acesse as documentações do [shapely](https://shapely.readthedocs.io/en/stable/index.html) e do [geopandas](http://geopandas.org/geometric_manipulations.html). 

## 5.1 Touches 
<hr style="border:1px solid #0077b9;">

O método touches() testa se a fronteira (boundary) de um objeto geográfico toca a fronteira de outro objeto. Utilizando a touches() pode-se, por exemplo, obter os municípios vizinhos a Altamira. O método retorna uma Series do Pandas do tipo bool.


```python
altamira_gdf = municipios_gdf[municipios_gdf["nm_municip"] == 'ALTAMIRA']

altamira_gdf.reset_index(drop=True, inplace=True)

touches_altamira_ps = municipios_gdf.touches(altamira_gdf.loc[0, 'geom'])

touches_altamira_ps.head()
```

 O retorno pode ser utilizado como uma máscara para acessar somente os vizinhos de altamira.

```python
vizinhos_altamira_gdf = municipios_gdf[touches_altamira_ps]

vizinhos_altamira_gdf.plot(edgecolor='black',figsize=(5, 5))
```

## 5.2 Union 
<hr style="border:1px solid #0077b9;">

O método de agregação unary_union realiza a união de um conjunto de objetos geométricos. Por exemplo, pode-se realizar a união dos vizinhos de Altamira da seguinte forma:

```python
vizinhos_union = vizinhos_altamira_gdf.unary_union

type(vizinhos_union)
```

Como o resultado do método não é um GeoSeries, para vizualizar o resultado através do .plot() devemos transformar o dado em um GeoSeries, da seguinte forma:

```python
vizinhos_union_gdf = gpd.GeoSeries(vizinhos_union)
vizinhos_union_gdf.crs = {'init': 'epsg:4674'}
vizinhos_union_gdf.plot(edgecolor='black',figsize=(5, 5))
```

Outra forma de aplicar a união entre objetos geométricos é através do método union().

```python
regiao_altamira_gdf = altamira_gdf.union(vizinhos_union_gdf)

regiao_altamira_gdf.plot(edgecolor='black',figsize=(5, 5))
```

## 5.3 Within 
<hr style="border:1px solid #0077b9;">

O método within() testa se um objeto geométrico se encontra dentro de outro. Através deste método podemos, por exemplo, obter todos os focos de incêndio que ocorreram no município de Altamira.  

```python
focos_within_altamira = focos_2017_gdf.within(altamira_gdf.loc[0, 'geom'])

focos_altamira_2017_gds = focos_2017_gdf[focos_within_altamira]

base = altamira_gdf.plot(edgecolor='black', figsize=(5, 5))

focos_altamira_2017_gds.plot(ax=base, color='red',  markersize=0.5)
```

## 5.4 Contains
<hr style="border:1px solid #0077b9;">
 
 O método contains é o inverso do método within().

```python
import pandas as pd
import numpy as np
regiao_focos_gs = pd.Series(np.empty((len(focos_2017_gdf)), dtype=bool) )
i = 0

for f in focos_2017_gdf.geometry:
    regiao_focos_gs[i] = regiao_altamira_gdf.contains(f).values[0]
    i = i + 1
    
focos_regiao_2017_gdf = focos_2017_gdf[regiao_focos_gs]

base = regiao_altamira_gdf.plot(edgecolor='black', figsize=(5, 5))

focos_regiao_2017_gdf.plot(ax=base, color='red',  markersize=0.5)
```

## 6.1 Focos por mês
<hr style="border:1px solid #0077b9;">
Primeiramente iremos alterar o formato da coluna data_obser para o tipo datetime para facilitar a manipulação dos dados desta coluna:

```python
focos_2017_gdf["data_obser"] = pd.to_datetime(focos_2017_gdf["data_obser"])
focos_2017_gdf.dtypes
```

```python
focos_2017_mes = focos_2017_gdf.groupby(focos_2017_gdf.data_obser.dt.month).uf.count()
focos_2017_mes
```

O objeto focos_mes retornado na operação acima corresponde a um pandas.core.series.Series:

```python
type(focos_2017_mes)
```

Na saída acima podemos notar o seguinte:

O nome da série é estado, por conta da coluna usada para realizar a contagem.
Os índices da série correspondem aos índices numéricos dos meses do ano.
Podemos re-indexar a série pelo nome do mês do ano. Para isso, podemos construir uma função lambda e aplicá-la através do operador map a cada um dos valores numéricos do índice do ano. Para transformar o mês do ano em um nome, utilizaremos o módulo calendar da Biblioteca Padrão Python.

```python
import calendar

novo_indice = map(lambda v : calendar.month_abbr[v], focos_2017_mes.index)

l = list(novo_indice)

l
```

Agora podemos construir explicitamente a nova série através do construtor pd.Series:

```python
focos_2017_mes = pd.Series(data=focos_2017_mes.values, index=l)
```

Podemos também ajustar o nome da série e o rótulo do índice:

```python
focos_2017_mes.name= "Número Focos/Mês"

focos_2017_mes.index.name="mes"
```

Podemos apresentar um gráfico de barras com o total de focos por mês:

```python
ax  = focos_2017_mes.plot(kind="bar", legend=False, fontsize=20, figsize=(20,10));
ax.set_title("Focos Mensal - 2017", fontsize=36);
ax.set_xlabel("Mes", fontsize=24);
ax.set_ylabel("#Focos", fontsize=24);
# ax.legend(loc=2, prop={'size': 20});
```
